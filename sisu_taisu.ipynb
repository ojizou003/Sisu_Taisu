§２. 爆発的な増加をグラフにしてみよう

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1, 10, 100)
y = 2 ** x
```

```python
plt.plot(x, y, label='$y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y = 2^x$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.linspace(1, 10, 100)
y = 1/2 ** x
```

```python
plt.plot(x, y, label='$y = (\\frac{1}{2})^x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y = (\\frac{1}{2})^x$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 60, 600)
y = (9/10) ** x
```

```python
plt.plot(x, y, label = '$y = (\\frac{9}{10})^x$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

§2.計算をラクにするために生まれた対数

```python
np.log2(8)
```

```python
x = np.linspace(0.1, 10, 100)
y = np.log2(x)
```

```python
plt.plot(x, y, label='$y=\log_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y1 = 2 ** x
y2 = x
y3 = np.log2(x)
```

```python
plt.plot(x, y1, label='$y=2^x$')
plt.plot(x, y2, label='$y=x$')
plt.plot(x, y3, label='$y=\log_2x$')
plt.ylim(-5, 5)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=1)  # x軸
plt.axvline(0, color='black', linewidth=1)  # y軸
plt.show()
```

対数グラフ

```python
x = np.linspace(0, 10, 100)
y = 2 ** x
```

```python
plt.plot(x, y, label='$y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.yscale('log')
plt.title('$y = 2^x$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot(x, y1, label='$y=2^x$')
plt.plot(x, y2, label= '$y=3^x$')
plt.plot(x, y3, label= '$y=4^x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.ylim(0, 5)
plt.grid(True)
plt.legend()
plt.show()
```
